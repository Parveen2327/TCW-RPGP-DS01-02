```python
num_cols = ['ssc_percentage', 'hsc_percentage', 'degree_percentage', 'emp_test_percentage', 'mba_percent']
plt.figure(figsize=(14, 8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='gender', hue='status', palette='Set2')
plt.title("Placement Status by Gender")
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.countplot(data=df, x='work_experience', hue='status', palette='coolwarm')
plt.title("Placement Status by Work Experience")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
numeric_df = df[num_cols].copy()
numeric_df['status'] = df['status']
sns.pairplot(numeric_df, hue='status', palette='husl', diag_kind='hist')
plt.show()
```
