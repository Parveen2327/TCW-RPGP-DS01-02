```python
def sample_non_numeric_values(df, col_name, num_samples=20):
    """Samples and returns non-numeric values from a specified column."""
    non_numeric_values = df[pd.to_numeric(df[col_name], errors='coerce').isna()][col_name].unique()
    return non_numeric_values[:num_samples]

# List of columns to check for non-numeric values (excluding 'Ticker')
numeric_cols = df.columns.drop('Ticker')

# Sample and display non-numeric values for each numeric column
for col in numeric_cols:
    non_numeric = sample_non_numeric_values(df, col)
    if len(non_numeric) > 0:
        print(f"Non-numeric values in column '{col}': {non_numeric}")
```

```python


# Melt the DataFrame to long format, selecting columns that represent stock volume
volume_cols = [col for col in df.columns if col.endswith('.4')]
id_vars = ['Ticker'] + [col for col in df.columns if not col.endswith('.4') and col != 'Ticker']
df_melted = df.melt(id_vars=id_vars, value_vars=volume_cols, var_name='Vaccine_Volume_Col', value_name='Quantity')

# Extract vaccine name from the melted column name
df_melted['Vaccine'] = df_melted['Vaccine_Volume_Col'].apply(lambda x: x.split('.')[0])

# Convert 'Quantity' to numeric, coercing errors
df_melted['Quantity'] = pd.to_numeric(df_melted['Quantity'], errors='coerce')

# Drop rows where Quantity is NaN after coercion
df_melted.dropna(subset=['Quantity'], inplace=True)

# Group by Vaccine and sum the Quantity
grouped_data = df_melted.groupby('Vaccine')['Quantity'].sum().reset_index()

plt.figure(figsize=(10,6))
# Create the bar plot using matplotlib.pyplot.bar
plt.bar(grouped_data['Vaccine'], grouped_data['Quantity'], color=sns.color_palette("viridis", len(grouped_data['Vaccine'])))
plt.title("Total Stock per Vaccine", fontsize=16)
plt.xticks(rotation=45)
plt.ylabel("Total Quantity")
plt.xlabel("Vaccine Type")
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

```python
# Melt the DataFrame to long format, selecting columns that represent stock volume
volume_cols = [col for col in df.columns if col.endswith('.4')]
id_vars = ['Ticker'] + [col for col in df.columns if not col.endswith('.4') and col != 'Ticker']
df_melted = df.melt(id_vars=id_vars, value_vars=volume_cols, var_name='Vaccine_Volume_Col', value_name='Quantity')

# Extract vaccine name from the melted column name
df_melted['Vaccine'] = df_melted['Vaccine_Volume_Col'].apply(lambda x: x.split('.')[0])

# Convert 'Quantity' to numeric, coercing errors
df_melted['Quantity'] = pd.to_numeric(df_melted['Quantity'], errors='coerce')

# Drop rows where Quantity is NaN after coercion
df_melted.dropna(subset=['Quantity'], inplace=True)

# Convert Ticker column to datetime format
df_melted['Ticker'] = pd.to_datetime(df_melted['Ticker'], errors='coerce')

# Drop rows where Ticker is NaN after coercion
df_melted.dropna(subset=['Ticker'], inplace=True)

# Sort by date
df_melted = df_melted.sort_values('Ticker')

# Plot time series
plt.figure(figsize=(12,6))
sns.lineplot(x="Ticker", y="Quantity", data=df_melted, marker="o", hue="Vaccine", palette="tab10")
plt.title("Vaccine Stock Over Time", fontsize=16)
plt.ylabel("Stock Quantity")
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.legend(title="Vaccine Type")
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```
